```python
'''
https://archive.ics.uci.edu/ml/datasets/Balance+Scale

Attribute Information:

1. Class Name: 3 (L, B, R) 
2. Left-Weight: 5 (1, 2, 3, 4, 5) 
3. Left-Distance: 5 (1, 2, 3, 4, 5) 
4. Right-Weight: 5 (1, 2, 3, 4, 5) 
5. Right-Distance: 5 (1, 2, 3, 4, 5)

Number of Instances: 625
L == 1
B == 2
R == 3
'''
import numpy as np
```

```python
# x = entrada
# y = saida

x = np.genfromtxt('DataSets/balance-scale.data', delimiter=',', usecols=(1,2,3,4))
y = np.genfromtxt('DataSets/balance-scale.data', delimiter=',', usecols=(0))
    
print(len(x))
print(len(y))
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
print(len(x_train))
x_train
```

```python
print(len(x_test))
x_test
```

```python
x_test[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier as knc
knn = knc(n_neighbors=17,p=2)
knn.fit(x_train, y_train)
labels = knn.predict(x_test)
print(len(labels))
labels
```

```python
y_test
```

```python
np.sum(labels == y_test) # comparação
(labels == y_test).sum() # comparação
```

```python
100 * np.sum(labels == y_test) / len(x_test) # porcentagem de acertos
```
